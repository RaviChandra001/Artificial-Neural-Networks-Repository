# **Exp 2 : Activation Functions**

```
def BinaryStep(x):
  return np.heaviside(x,1)

x = np.linspace(-10,10)
plt.plot(x,BinaryStep(x))
plt.title("Binary Step")
plt.show()
```

Linear Activation


```
def LinearAct(x):
  return x

x = np.linspace(-10,10)
plt.plot(x,LinearAct(x))
plt.title("Linear Step")
plt.show()
```

Sigmoid Activation

```
def SigmoidAct(x):
  return (1/(1+np.exp(-x)))

x = np.linspace(-10,10)
plt.plot(x,SigmoidAct(x))
plt.title("Sigmoid Step")
plt.show()
```

Hyperbolic TanH

```
def TanH(x):
  return np.tanh(x)

x = np.linspace(-10,10)
plt.plot(x,TanH(x))
plt.title("Hyperbolic Step")
plt.show()
```

RelU function

```
def ReLuAct(x):
  x1 = []
  for i in x:
    if(i<0):
      x1.append(0)
    else:
      x1.append(i)
  return x1

x = np.linspace(-10,10)
plt.plot(x,ReLuAct(x))
plt.title("ReLu fun Step")
plt.show()
```

Softmax Activation fun


```
def Softmax(x):
  return np.exp(x)/np.sum(np.exp(x))

x = np.linspace(-10,10)
plt.plot(x,Softmax(x))
plt.title("Softmax Step Function")
plt.show()
```

# ** Exp 3 : Program to implement the precptron For Logical AND**

```
# Taking inputs
x = np.array([0,0])

# Weights
w = np.array([1,1])

# Bias
b = -1.5

# Net Input
Yin = np.dot(w,x)+b
print(Yin)

# Activation Unit Step
if(Yin >= 0):
  print('1')
else:
  print(0)

# By the function
def PercAnd(x1,x2,w1,w2,b):
  x = np.array([x1,x2])
  w = np.array([w1,w2])

  Yin = np.dot(x,w) + b
  if(Yin >  0 ):
    print('1')
  else:
    print('0')

  # Function calls
PercAnd(0,0,1,1,-1.5)

PercAnd(0,1,1,1,-1.5)

PercAnd(1,0,1,1,-1.5)

PercAnd(1,1,1,1,-1.5)
```

# **Program to implement the precptron For Logical OR**

```
# Inputs
x = np.array([0,0])

# Weights
w = np.array([3,3])

# Bias
b = -1.5

# Net Input
Yin = np.dot(w,x)+b
print(Yin)

# Activation Unit Step
if(Yin>=1):
  print("1")
else:
  print('0')


# User defined function with binary step as activation

def PercOR(x1,x2,w1,w2,b):
  x=np.array([x1,x2])
  w=np.array([w1,w2])
  yin=np.dot(w,x)+b
  if(yin>=0):
    return 1
  else:
    return 0


PercOR(0,0,3,3,-1.5)

PercOR(0,1,3,3,-1.5)

PercOR(1,0,3,3,-1.5)

PercOR(1,1,3,3,-1.5)
```

# **Exp 5 : Implementing Find-S Algorithm in Machine Learning**

```
#  Data Feeding
Data={
    "weather":["sunny","sunny","rainy","sunny",],
    "humidity":["high","low","low","low",],
    "air":["warm","warm","warm","cool",],
    "play":["yes","yes","no","yes",],
}

df = pd.DataFrame(Data)

df.head()

def FindS(df, ho=["","","","",]):

  # Check the example is positive or not
    df = df[df['play'] != 'no']

  # Copy the positive example from DF into the ho
    ho = df.iloc[0].copy()

  # Iterating the DataFrame to compair values with ho

    for i in range(1, len(df)):
        if all(df.iloc[i] == ho):
            pass

        else:
  # If change in example value replace it with the Most specific "?"
            for j in range(len(ho)):
                if ho[j] != df.iloc[i][j]:
                    ho[j] = "?"

    return ho


result = FindS(df)

# Check if ho having a Most Specific hypothesis or not
if result is not None:
    print("Most specific hypothesis is:")
    print(result)
else:
    print("No positive examples found.")
```

# **Exp 6 : Candidate Elimination ALgo**

```
import numpy as np
import pandas as pd

data = pd.DataFrame(data=pd.read_csv('enjoysport.csv'))
concepts = np.array(data.iloc[:, 0:-1])
print(concepts)
target = np.array(data.iloc[:, -1])
print(target)

def learn(concepts, target):
    specific_h = concepts[0].copy()
    print("Initialization of specific_h and general_h")
    print(specific_h)
    general_h = [["?" for _ in range(len(specific_h))] for _ in range(len(specific_h))]
    print(general_h)
    for i, h in enumerate(concepts):
        if target[i] == "yes":
            for x in range(len(specific_h)):
                if h[x] != specific_h[x]:
                    specific_h[x] = '?'
                    general_h[x][x] = '?'
            print(specific_h)
            print(general_h)
        if target[i] == "no":
            for x in range(len(specific_h)):
                if h[x] != specific_h[x]:
                    general_h[x][x] = specific_h[x]
                else:
                    general_h[x][x] = '?'
            print("Steps of Candidate Elimination Algorithm", i + 1)
            print(specific_h)
            print(general_h)
    indices = [i for i, val in enumerate(general_h) if val == ['?', '?', '?', '?', '?', '?']]
    for i in indices:
        general_h.remove(['?', '?', '?', '?', '?', '?'])
    return specific_h, general_h

s_final, g_final = learn(concepts, target)
print("Final Specific_h:", s_final, sep="\n")
print("Final General_h:", g_final, sep="\n")
```

# Exp : Self Organization Map


```
import numpy as np

# Create Weight Matrix
w = [[0.2, 0.6, 0.4, 0.9, 0.2], [0.3, 0.5, 0.7, 0.6, 0.8]]
w = np.array(w)
print(w)

# Create A list of hold vector values
x = [0.3, 0.4]

# Create Variable For learning rate
a = 0.3

# Calculate Euclidean Distance for each neuron from vectors
d = np.array([0.0] * len(w[0]))
for i in range(len(w)):
    for j in range(len(w[0])):
        d[i] += (w[i][j] - x[i]) ** 2

print(d)

# Find the cluster with the minimum distance from the vector

minv = max(d)
minI = 0

for i in range(len(d)):
    if (minv > d[i]):
        minV = d[i]
        minI = i + 1

print('Winning Cluster is C', minI,'with Distance ',minV)

# Update Weights for winning cluster & its neighborhood
for i in range(len(w)):
    for j in range(minI - 2, minI + 1):
        if j >= 0 and j < len(w[0]):
            w[i][j] += a * (x[i] - w[i][j])  # Corrected this line

print('Updated Weights Matrix is : ')
print(w)
```

# **Exp 8 : Implementing Naive Bayes Algo using SkLearn**

```
# load the iris dataset
from sklearn.datasets import load_iris
iris = load_iris()

# store the feature matrix (X) and response vector (y)
X = iris.data
y = iris.target

# splitting X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)

# training the model on training set
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# making predictions on the testing set
y_pred = gnb.predict(X_test)

# comparing actual response values (y_test) with predicted response values (y_pred)
from sklearn import metrics
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
```

# **Exp 9 : KNN (K-Nearest Neighbour)**

```
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)

# Define the kNN classifier
k = 3  # Number of neighbors
knn = KNeighborsClassifier(n_neighbors=k)

# Train the classifier
knn.fit(X_train, y_train)

# Predict the labels for the test set
y_pred = knn.predict(X_test)

# Calculate the accuracy of the classifier
accuracy = metrics.accuracy_score(y_test, y_pred)
print("k-Nearest Neighbors (kNN) model accuracy (in %):", accuracy * 100)
```
